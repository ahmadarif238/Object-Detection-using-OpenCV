```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = "CV/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model = "CV/frozen_inference_graph.pb"
```

```python
model =cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = []
file_name = "CV/labels.txt"
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True)
```

```python
img = cv2.imread('boy.jpg')
plt.imshow(img)
```

```python
ClassIndex, confidece, bbox = model.detect(img, confThreshold= 0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(),bbox):
    cv2.rectangle(img, boxes, (255,0,0), 2)
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale = font_scale, color =(0,255,0), thickness= 3) 
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# Detection using my webcam 
cap = cv2.VideoCapture('Pedestrians jaywalking - City of Vancouver Road User Safety.mp4')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError('Cannot open the video')

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Perform detection
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    print(ClassIndex)

    # Draw bounding boxes if detections are made
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:  # Assuming class IDs are within a certain range
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                label = f'{classLabels[ClassInd-1]}: {conf:.2f}'
                cv2.putText(frame, label, (boxes[0], boxes[1]-10), font, font_scale, (0, 255, 0), 3)

    cv2.imshow('Object detection by Arif', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Detection using a video
video_path = 'Pedestrians jaywalking -  City of Vancouver Road User Safety.mp4'
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise IOError('Cannot open the video file')

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Perform detection
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    # Draw bounding boxes if detections are made
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:  # Assuming class IDs are within a certain range
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                label = f'{classLabels[ClassInd-1]}: {conf:.2f}'
                cv2.putText(frame, label, (boxes[0], boxes[1]-10), font, font_scale, (0, 255, 0), 3)

    cv2.imshow('Object detection by Arif', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
